```python
import pandas as pd
import numpy as np
df=pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv')
df
df.head()

```

```python
df_features=df[['Movie_Genre','Movie_Keywords','Movie_Tagline','Movie_Cast','Movie_Director']].fillna('')
df_features
```

```python
x_test=df_features['Movie_Genre']+df_features['Movie_Keywords']+df_features['Movie_Tagline']+df_features['Movie_Cast']+df_features['Movie_Director']
x_test
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x_tfidf=tfidf.fit_transform(x_test)
print(x_tfidf)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(x_tfidf)
similarity
```

```python
fav_movie=input('Enter your favourite movie name:')
```

```python
all_movies=df['Movie_Title'].tolist()
import difflib
movie_close_match=difflib.get_close_matches(fav_movie,all_movies)
print(movie_close_match)
```

```python
close_match=movie_close_match[0]
print(close_match)
```

```python
index_of_close_match=df[df.Movie_Title==close_match]['Movie_ID'].values[0]
print(index_of_close_match)
```

```python
recommendation_score=list(enumerate(similarity[index_of_close_match]))
print(recommendation_score)
```

```python
len(recommendation_score)
```

```python
sorted_similar_movies=sorted(recommendation_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
print("similar movies")
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=df[df.index==index]['Movie_Title'].values[0]
  if(i<31):
    print(i,'.',title_from_index)
    i+=1
```
